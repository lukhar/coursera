# Non personalized recommenders

```python
import pandas as pd
```

```python
movie_ratings = pd.read_csv('movie_data.csv', index_col=0)
```

```python
movies = {int(ident): title for ident, title in  
          [tuple(movie_name.split(':', 1)) 
           for movie_name in movie_ratings.columns]}

movie_ratings.rename(columns=lambda name: tuple(name.split(':', 1))[0], inplace=True)

pd.DataFrame(movies.items(), columns=['id','title'])
```

```python
movie_ratings
```

## Mean rate ranking

```python
mean_ranking = movie_ratings.mean()
mean_ranking.sort(ascending=False)
mean_ranking.head(5)
```

## Rate count ranking

```python
count_ranking = movie_ratings.count()
count_ranking.sort(ascending=False)
count_ranking.head(5)
```

## Positive ratio ranking

```python
positive_ratio_ranking = movie_ratings[movie_ratings[:] >= 4].count() / movie_ratings.count()
positive_ratio_ranking.sort(ascending=False)
positive_ratio_ranking.head(5)
```

```python
p_star_wars = len(movie_ratings[movie_ratings['260'] >= 0]) / float(len(movie_ratings))

ranking = []
for movie in movie_ratings.columns:
    p_rated_with_star_wars = len(movie_ratings[(movie_ratings['260'] >=0) & (movie_ratings[movie] >=0)]) / float(len(movie_ratings))
    ranking.append((movie, p_rated_with_star_wars/p_star_wars))
    
# star_wars discounted
top_five = sorted(ranking, key=lambda rank: rank[1], reverse=True)[1:6]

for movie, rate in top_five:
    print '{}\t{:f}'.format(movie, rate)
```
